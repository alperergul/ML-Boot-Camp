### Matplotlib example workflow

```python
# 0. import matplotlib

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data

x = [1,2,3,4]
y = [12,20,10,30]

# 2. Setup plot

fig, ax = plt.subplots(figsize=(10,10)) #width height

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot

ax.set(title="Test title", xlabel="x-label", ylabel="y-label")

# 5. Save and show
plt.savefig("images/sample-plot.png")
# plt.show()
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
# PLot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(x);
```

```python
# Subplot option 2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
hearth_disease = pd.read_csv("heart-disease.csv")
hearth_disease
```

```python
# Create a histogram of age
hearth_disease["age"].plot.hist(bins=50)
```

```python
hearth_disease.head()
```

```python
hearth_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50 = hearth_disease[hearth_disease["age"] > 50 ]
over_50.head()
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c="target", figsize=(10,6));
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
# ax.set_xlim([45,100])
```

```python
## OO method from scratch

fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"],);

ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
# Subplot of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"])
ax0.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="target")
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

scatter= ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
ax1.set(title="Heart Disease and Thalach Levels", xlabel="Age", ylabel="Thalach")
ax1.legend(*scatter.legend_elements(), title="target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

fig.suptitle("Hearth Disease Analysis", fontsize=16, fontweight="bold");
```
